```python
#4 Write a Python program to extract all the URLs from the webpage python.org that are nested w ithin <li> tags from
import requests
import re
from bs4 import BeautifulSoup as bs
def geturl(html):
    pattern = re.compile(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+')
   
    urllist=re.findall(pattern,html)
    return(urllist)
url = "https://www.python.org/"
page = requests.get(url).text
soup = bs(page,"html.parser")
lis = soup.find_all('li')
lis
geturl(str(lis))
```

```python
#5 Write a Python program to extract all the text from a given web page.
import requests
from bs4 import BeautifulSoup
url = 'https://www.python.org/'
reqs = requests.get(url).text
soup = BeautifulSoup(reqs, 'lxml')
print("Text from the said page:")
print(soup.get_text())

```

```python
#6 Write a Python program to retrieve the HTML code of the title, its text, and the HTML code of its parent.
import requests
from bs4 import BeautifulSoup
url = 'https://en.wikipedia.org/wiki/Wiki'
reqs = requests.get(url).text
soup = BeautifulSoup(reqs,'html.parser')

print("title:\n",soup.title)
print("text:\n",soup.title.text)
print("parent:\n",soup.title.parent)

```

```python
#7 Write a Python program to create a Beautiful Soup parse tree into a nicely formatted Unicode string, with a separate line for each HTML/XML tag and string.
import requests
from bs4 import BeautifulSoup
url = 'https://www.python.org/'
reqs = requests.get(url).text
soup = BeautifulSoup(reqs,'html.parser')
print(soup.prettify())
```

```python
#8 Write a Python program to remove the contents of a tag in a given html document.
import requests
from bs4 import BeautifulSoup
url = 'https://www.python.org/'
reqs = requests.get(url).text
soup = BeautifulSoup(reqs,'html.parser')
soup.a.clear()
print(soup.a)
```

```python
#9 Write a Python program to wrap an element in the specified tag and create the new wrapper.
from bs4 import BeautifulSoup
soup = BeautifulSoup("<p>Python exercises.</p>", "lxml")
print("Original Markup:",soup)
soup.p.string.wrap(soup.new_tag("i"))
print("New Markup:",soup)
```

```python
#10 Write a Python program to remove a tag from a given tree of html document and destroy it and its contents.
import requests
from bs4 import BeautifulSoup
url = 'https://www.python.org/'
reqs = requests.get(url).text
soup = BeautifulSoup(reqs,'html.parser')
tag1 = soup.a
print("Before delete a tag: ",tag1)
tag2 = soup.a.decompose()
print("After delete a tag: ",tag2)
```
